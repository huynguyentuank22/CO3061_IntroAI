```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('sudoku_results.csv')
df
```

```python
df.info()
```

```python
# Set seaborn style
sns.set(style="whitegrid")

# Create plots
fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# Execution Time vs. Difficulty
sns.boxplot(data=df, x="Difficulty", y="Time (s)", hue="Algorithm", ax=axes[0, 0])
axes[0, 0].set_title("Execution Time vs. Difficulty Level")

# Memory Usage Comparison
sns.boxplot(data=df, x="Difficulty", y="Memory (KB)", hue="Algorithm", ax=axes[0, 1])
axes[0, 1].set_title("Memory Usage vs. Difficulty Level")

# Number of Steps vs. Difficulty
sns.boxplot(data=df, x="Difficulty", y="Steps", hue="Algorithm", ax=axes[1, 0])
axes[1, 0].set_title("Number of Steps vs. Difficulty Level")

# Backtracks for DFS only
df_dfs = df[df["Algorithm"] == "DFS"]
sns.boxplot(data=df_dfs, x="Difficulty", y="Backtracks", ax=axes[1, 1])
axes[1, 1].set_title("Backtracks vs. Difficulty Level (DFS Only)")

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
diffs = ["Easy", "Medium", "Hard", "Expert"]
plt.figure(figsize=(10, 6))
sns.boxplot(x='Difficulty', y='Time (s)', hue='Algorithm', data=df, order=diffs)
    
plt.title("Execution Time vs Difficulty Level for A* and DFS")
plt.xlabel("Difficulty Level")  
plt.ylabel("Execution Time (seconds)")
plt.legend(title="Algorithm")
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
